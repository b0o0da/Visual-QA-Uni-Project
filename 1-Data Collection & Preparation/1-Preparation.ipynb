```python
import pandas as pd

files = [
    "visual_qa_batch1_rows0-49.csv",
    "visual_qa_batch2_rows50-99.csv",
    "visual_qa_batch3_rows100-149.csv",
    "visual_qa_batch4_rows150-199.csv",
    "visual_qa_rows200_to_end.csv"
    
]

dfs = []

for file in files:
    temp = pd.read_csv(file)
    dfs.append(temp)

final_df = pd.concat(dfs, ignore_index=True)
final_df = final_df.drop_duplicates().reset_index(drop=True)
final_df = final_df.drop(columns=["question_num"], errors="ignore")

final_df.to_csv("visual_qa_all_concatenated.csv", index=False)

print("Done!")
print("Total rows:", len(final_df))
print("Columns:", final_df.columns.tolist())
final_df
```

```python
['image_path', 'image_url', 'source', 'topic', 'caption', 'question', 'answer']
['image_id', 'caption', 'question', 'answer', 'source', 'image_url', 'topic']

df_GPT = pd.read_csv('high_quality_visual_qa.csv')

df_GPT["image_path"] = final_df["image_path"]
df_GPT.drop(columns=["image_id"], inplace=True, errors="ignore")
df_GPT["source"] = final_df["source"]
df_GPT["image_url"] = final_df["image_url"]
df_GPT["topic"] = final_df["topic"]
```

```python
import pandas as pd

file1 = "visual_qa_all_concatenated.csv"

df1 = pd.read_csv(file1)

df2 = df_GPT.copy()

final_cols = [
    "image_path",
    "image_url",
    "source",
    "topic",
    "caption",
    "question",
    "answer"
]

for col in final_cols:
    if col not in df1.columns:
        df1[col] = ""

    if col not in df2.columns:
        df2[col] = ""

df1 = df1[final_cols]
df2 = df2[final_cols]

final_df = pd.concat([df1, df2], ignore_index=True)

final_df.to_csv("final_visual_qa_dataset.csv", index=False)

print("Done!")
print(final_df.shape)
print(final_df.columns.tolist())
final_df
```
